```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_used_cars=pd.DataFrame([["audi",419],
                           ["bmw",636],
                           ["mercedics-benz",820],
                           ["mitsubiha",306],
                           ["renault",438],
                           ["toyta",509],
                           ["volkswagen",875]
                          ],columns=["brand","car listings"])
```

```python
df_used_cars
```

```python
# bar plot
plt.figure(figsize=(10,10))
plt.bar(x=df_used_cars["brand"],height=df_used_cars["car listings"])
plt.title("cars types",fontsize=20)
plt.xlabel("brand")
plt.ylabel("car numbers")
plt.show()
```

```python
df_engine_fuel=pd.DataFrame([
                            ["diesel",2019],
                           ["gas",613],
                           ["others",154],
                           ["petrol",1559]
                            ],columns=["engine fuel type","number of cars"])
```

```python
df_engine_fuel
```

```python
#pie chart
plt.figure(figsize=(10,10))
plt.pie(x=df_engine_fuel["number of cars"],labels=df_engine_fuel["engine fuel type"])
plt.show()
```
