```python
# how remove outlier
def remove_outlier(df):
    q1=df.quantile(.25) # q1
    q3=df.quantile(.75) # q1
    iqr=q3-q1
    df[(df<(q1-1.5*iqr))|(q3+1.5*iqr)]=np.nan
    df.dropna()
    return df
df=remove_outlier(df)
```
